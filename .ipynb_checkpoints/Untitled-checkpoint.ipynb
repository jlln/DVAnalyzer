```python
import pandas
import matplotlib.pyplot as plt
import seaborn
import matplotlib as mpl
%matplotlib inline
```

```python
seg_data = pandas.read_csv("segmentation_comparison.txt")
```

```python
seg_data.head()
```

```python
seg_data["snr"] = (seg_data.f - seg_data.b)/seg_data.n
```

```python
f, ax = plt.subplots(figsize=(7, 10))
seaborn.regplot("snr","four_means_ce",seg_data,fit_reg=False)
```

```python
segm = pandas.melt(seg_data,id_vars=["b","f","n","snr"])
```

```python
segm['measurement_type'] = ["pe" if "pe" in x else "ce" for x in segm.variable.values ]
segm['segmentation_method'] = [x[:-3] for x in segm.variable.values]
```

```python
segm.head()
```

```python
ce = segm[segm.measurement_type=="ce"]
pe = segm[segm.measurement_type=="pe"]
```

```python
from ggplot import *
```

```python

ggplot(ce,aes(x="snr",y="value",color="segmentation_method")) + stat_smooth()
```

```python

ggplot(pe,aes(x="snr",y="value",color="segmentation_method")) + stat_smooth()
```

```python
import numpy as np
```

```python
abs_obs = np.abs(segm.value)
```

```python
segma = segm
segma['abs_value'] = abs_obs
ce = segma[segma.measurement_type=="ce"]
pe = segma[segma.measurement_type=="pe"]
```

```python
ggplot(ce,aes(x="snr",y="abs_value",color="segmentation_method")) + geom_point()
```

```python
sct_fm=segma[(segma.segmentation_method=="SCT") | (segma.segmentation_method =="four_means")]
```

```python
ggplot(sct_fm,aes(x="snr",y="abs_value")) + geom_point() + facet_grid("measurement_type","segmentation_method")
```

```python
segsort = segma.sort(columns=["variable","snr"])
```

```python
ggplot(segsort,aes(x="snr",y="abs_value",color="segmentation_method")) + geom_point()+facet_grid("segmentation_method","measurement_type") 
```

```python

ggplot(segsort,aes(x="snr",y="abs_value",color="segmentation_method")) + stat_smooth() + facet_grid("segmentation_method","measurement_type")
```

```python
segsort.head()
```

```python
segsort.columns = ["b","f","n","Signal to Noise Ratio","variable","value","Measurement Error Metric","Segmentation Method","Absolute Value"]
```

```python
segsort.loc[segsort["Measurement Error Metric"] == "ce","Measurement Error Metric"] = "Object Count Error"
segsort.loc[segsort["Measurement Error Metric"] == "pe","Measurement Error Metric"] = "Foreground Pixel Count Error"
```

```python

ggplot(segsort,aes(x="Signal to Noise Ratio",y="Absolute Value",color="Segmentation Method")) + geom_point() + facet_grid("Segmentation Method","Measurement Error Metric",scales="freey")
```
